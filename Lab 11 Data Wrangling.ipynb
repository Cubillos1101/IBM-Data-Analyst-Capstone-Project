### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
df.head()
```

<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
df_summarize = pd.DataFrame({
    'Column': df.columns,
    'DataType': df.dtypes.values,
    'Unique Values': [df[col].nunique() for col in df.columns],
    'Missing Values': df.isnull().sum().values,
    'Non-Missing Values': df.notnull().sum().values
})
df_summarize.set_index('Column', inplace=True)
df_summarize
```

```python
df.describe()
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
missing_values = df.isnull().sum()  
missing_percentage = (missing_values / len(df)) * 100 

missing_data = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percentage
})
high_missing_data = missing_data[missing_data['Percentage'] > 70]
df.drop(columns=high_missing_data.index, inplace=True)
```

```python
null_values = df['Country'].isnull().sum()
print("Valores nulos:", null_values)
```

```python
df = df[df['Country'].notnull()]
value_counts = df['Country'].value_counts()
more_than_100 = value_counts[value_counts >= 100]
print(more_than_100.sum())
filtered_df = df[df['Country'].isin(more_than_100.index)]
print(filtered_df)
```

<h5>3.2 Standardize entries in columns 'Employment' by mapping inconsistent values to a consistent format.</h5>


```python
def split_employment(row):
    if isinstance(row, str):         
        parts = row.split(';')
        jobs_times = [p.strip().split(',') for p in parts]
        
        job1, time1, job2, time2 = None, None, None, None
        if len(jobs_times) > 0:
            job1 = jobs_times[0][0].strip()
            time1 = jobs_times[0][1].strip() if len(jobs_times[0]) > 1 else None
        if len(jobs_times) > 1:
            job2 = jobs_times[1][0].strip()
            time2 = jobs_times[1][1].strip() if len(jobs_times[1]) > 1 else None
        
        return pd.Series([job1, time1, job2, time2])
    else:
        return pd.Series([None, None, None, None])

df[['Job1', 'Time1', 'Job2', 'Time2']] = df['Employment'].apply(split_employment)
df=df.drop(columns=['Employment'])
df.head()
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
pd.get_dummies(df,columns=['Job1','Time1','Job2','Time2'])
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
missing_values = df.isnull().sum()  
missing_percentage = (missing_values / len(df)) * 100 

missing_data = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percentage
})

missing_data = missing_data.sort_values(by='Missing Values', ascending=False)

missing_data.head(10)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].median())
df['ConvertedCompYearly'].isnull().sum()
df['ConvertedCompYearly'].head()
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
df['RemoteWork'] = df['RemoteWork'].fillna(df['RemoteWork'].mode()[0])
df['RemoteWork'].isnull().sum()
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
df['ConvertedCompYearly_MinMax'] = (df['ConvertedCompYearly'] - df['ConvertedCompYearly'].min()) / (df['ConvertedCompYearly'].max() - df['ConvertedCompYearly'].min())
df['ConvertedCompYearly_MinMax'].head()
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
df['ConvertedCompYearly_log'] = np.log(df['ConvertedCompYearly'])
df['ConvertedCompYearly_log'].skew()
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
df.dropna(subset=['YearsCodePro'], inplace=True)

df['YearsCodePro'] = df['YearsCodePro'].replace('Less than 1 year', 0)
df['YearsCodePro'] = df['YearsCodePro'].replace('More than 50 years', 51)

df['YearsCodePro'] = df['YearsCodePro'].astype(float)

def classify_experience(years):
    if years <= 2:
        return 'Junior'
    elif 3 <= years <= 5:
        return 'Mid-Level'
    elif 6 <= years <= 10:
        return 'Senior'
    else:
        return 'Expert'

df['ExperienceLevel'] = df['YearsCodePro'].apply(classify_experience)


df['ExperienceLevel'].head()
```
